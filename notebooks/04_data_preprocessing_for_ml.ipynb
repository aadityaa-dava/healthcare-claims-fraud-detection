# Healthcare Claims Fraud Detection

## Data Preprocessing for Machine Learning

This notebook prepares the provider-level dataset for machine learning by performing feature selection, train-test splitting, feature scaling, handling class imbalance using SMOTE, and creation of final modeling datasets.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings("ignore")
print("Libraries Imported Successfully")
```

## Load Provider-Level Dataset

```python
provider_df = pd.read_csv("../data/processed/provider_features.csv")
print("Dataset Loaded Successfully")
```

## Dataset Overview

```python
print("Shape:",provider_df.shape)
provider_df.info()
```

### Interpretation

All variables are complete with no missing values.

The dataset contains numerical features suitable for machine learning and a binary target variable representing provider fraud status.

## Feature Selection

```python
features_to_drop = ["Provider", "PotentialFraud", "Fraud"]
X = provider_df.drop( columns=features_to_drop)
y = provider_df["Fraud"]
print("Feature Matrix Shape:", X.shape)
print("Target Vector Shape:", y.shape)
```

A total of 17 predictor variables were selected for modeling.

Identifier columns and the original target label were removed to prevent data leakage while retaining all engineered provider-level features.

## Review Modeling Features

```python
X.columns.tolist()
```

### Interpretation

The selected features include provider utilization metrics, reimbursement measures, beneficiary characteristics, and healthcare activity indicators.

Several features identified during exploratory analysis as strong fraud indicators are included in the modeling dataset.

## Split Dataset into Training and Testing Sets

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

### Interpretation

The dataset was successfully divided into training and testing subsets using an 80-20 split.

- Training Set: 4,328 providers
- Testing Set: 1,082 providers

Stratified sampling was used to preserve the original fraud distribution in both datasets, ensuring representative model evaluation.

## Verify Class Distribution

```python
print("Training Set Distribution")
print(y_train.value_counts())
print("\nTraining Set Percentage")
print(round(y_train.value_counts(normalize=True) * 100, 2))
print("\nTesting Set Distribution")
print(y_test.value_counts())
print("\nTesting Set Percentage")
print(round(y_test.value_counts(normalize=True) * 100, 2))
```

The fraud distribution remains consistent across the training and testing datasets.

The use of stratified sampling successfully preserved the original class proportions, preventing sampling bias and ensuring reliable model evaluation.

## Standardize Numerical Features

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print("Scaling Completed Successfully")
```

Feature scaling was successfully applied using StandardScaler.

All numerical variables were transformed to have approximately zero mean and unit variance, which improves the performance of distance-based and linear machine learning algorithms.

## Verify Scaled Features

```python
pd.DataFrame(X_train_scaled, columns=X.columns).head()
```

### Interpretation

The transformed feature values are centered around zero and exhibit standardized ranges.

This confirms that scaling was correctly applied and that the dataset is ready for class balancing and machine learning model training.

## Handle Class Imbalance Using SMOTE

```python
print("Before SMOTE")
print(y_train.value_counts())
```

### Interpretation

The training dataset exhibits significant class imbalance.

Fraudulent providers represent only a small proportion of observations, which may cause machine learning models to become biased toward the majority class.

```python
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(
    X_train_scaled,
    y_train
)
print("SMOTE Applied Successfully")
```

## Verify Class Distribution After SMOTE

```python
print("After SMOTE")
print(pd.Series(y_train_smote).value_counts())
```

### Interpretation

SMOTE generated synthetic fraud-provider observations to balance the training dataset.

The resulting dataset contains an equal number of fraudulent and non-fraudulent providers, enabling machine learning models to learn fraud patterns more effectively.

## Create Final Modeling Datasets

```python
X_train_final = pd.DataFrame(X_train_smote, columns=X.columns)
X_test_final = pd.DataFrame(X_test_scaled, columns=X.columns)
y_train_final = pd.Series(y_train_smote, name="Fraud")
y_test_final = pd.Series(y_test, name="Fraud")
print("Final Modeling Datasets Created")
```

The final modeling datasets have been successfully created.

Training data includes balanced fraud classes produced through SMOTE, while the testing dataset remains unchanged to provide an unbiased evaluation of model performance.

## Verify Final Dataset Dimensions

```python
print("X_train_final:", X_train_final.shape)
print("X_test_final :", X_test_final.shape)
print("y_train_final:", y_train_final.shape)
print("y_test_final :", y_test_final.shape)
```

## Save Processed Modeling Datasets

```python
import os
os.makedirs("../data/modeling", exist_ok=True)
X_train_final.to_csv("../data/modeling/X_train_final.csv", index=False)
X_test_final.to_csv("../data/modeling/X_test_final.csv", index=False)
y_train_final.to_csv("../data/modeling/y_train_final.csv", index=False)
y_test_final.to_csv("../data/modeling/y_test_final.csv", index=False)
print("Modeling Datasets Saved Successfully")
```
