# Healthcare Claims Fraud Detection

## Fraud Detection Modeling

This notebook develops and evaluates machine learning models to identify potentially fraudulent healthcare providers. Multiple classification algorithms will be trained and compared using standard evaluation metrics to determine the most effective fraud detection model.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings("ignore")
```

## Load Modeling Datasets

```python
X_train = pd.read_csv("../data/modeling/X_train_final.csv")
X_test = pd.read_csv("../data/modeling/X_test_final.csv")
y_train = pd.read_csv("../data/modeling/y_train_final.csv")
y_test = pd.read_csv("../data/modeling/y_test_final.csv")
y_train = y_train.squeeze()
y_test = y_test.squeeze()
print("Datasets Loaded Successfully")
```

## Verify Dataset Dimensions

```python
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

## Logistic Regression Model

```python
log_reg = LogisticRegression(
    random_state=42,
    max_iter=1000
)
log_reg.fit(X_train, y_train)
```

### Interpretation

The Logistic Regression model was successfully trained using the balanced training dataset.

This model serves as a baseline classifier and will be used as a benchmark for evaluating more advanced algorithms such as Random Forest and XGBoost.

## Generate Predictions

```python
y_pred_lr = log_reg.predict(X_test)
y_prob_lr = log_reg.predict_proba(X_test)[:, 1]
print("Predictions Generated Successfully")
```

## Evaluate Logistic Regression Performance

```python
lr_accuracy = accuracy_score(y_test, y_pred_lr)
lr_precision = precision_score(y_test, y_pred_lr)
lr_recall = recall_score(y_test, y_pred_lr)
lr_f1 = f1_score(y_test, y_pred_lr)
lr_auc = roc_auc_score(y_test, y_prob_lr)
lr_results = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC"],
    "Value": [lr_accuracy, lr_precision, lr_recall, lr_f1, lr_auc]
})
lr_results
```

### Interpretation

The Logistic Regression model achieved strong overall performance in detecting fraudulent providers.

Key observations:

- Accuracy: 90.57%
- Recall: 91.09%
- ROC-AUC: 96.99%

The high recall indicates that the model successfully identifies most fraudulent providers, which is important in fraud detection because missing fraudulent cases can be costly.

Although precision is lower (49.73%), this is expected in fraud detection problems where maximizing fraud detection is generally prioritized over minimizing false alarms.

The ROC-AUC score of 0.97 demonstrates excellent class separation capability, indicating that the model effectively distinguishes fraudulent providers from non-fraudulent providers.

## Logistic Regression Confusion Matrix

```python
lr_cm = confusion_matrix(y_test, y_pred_lr)
cm_df = pd.DataFrame(lr_cm,
    index=["Actual Non-Fraud", "Actual Fraud"],
    columns=["Predicted Non-Fraud", "Predicted Fraud"])
cm_df
```

### Interpretation

The Logistic Regression model correctly classified the majority of providers.

Results:

- True Negatives (TN): 888
- False Positives (FP): 93
- False Negatives (FN): 9
- True Positives (TP): 92

The model missed only 9 fraudulent providers while successfully identifying 92 fraud cases.

This aligns with the high recall score (91.09%), indicating that the model is highly effective at detecting fraud. Although 93 non-fraud providers were incorrectly flagged as fraud, this tradeoff is generally acceptable in fraud detection applications where identifying fraudulent activity is the primary objective.

## Visualize Confusion Matrix

```python
fig = px.imshow(
    lr_cm,
    text_auto=True,
    color_continuous_scale="Blues",
    labels=dict(
        x="Predicted",
        y="Actual",
        color="Count"
    ),
    x=["Non-Fraud", "Fraud"],
    y=["Non-Fraud", "Fraud"],
    title="Logistic Regression Confusion Matrix"
)
fig.update_layout(
    width=700,
    height=600
)
fig.show()
```

### Interpretation

The confusion matrix visualization confirms the strong fraud-detection capability of the Logistic Regression model.

Most fraudulent providers were correctly identified, resulting in a very low number of false negatives. This is particularly important in healthcare fraud detection, where failing to detect fraud can lead to substantial financial losses.

While some non-fraud providers were incorrectly flagged as fraudulent, the model prioritizes fraud detection effectiveness over minimizing false alarms.

## Classification Report

```python
print(classification_report(y_test, y_pred_lr, target_names=["Non-Fraud", "Fraud"]))
```

### Interpretation

The Logistic Regression model demonstrates strong fraud detection performance.

For fraudulent providers:

- Precision: 50%
- Recall: 91%
- F1 Score: 64%

The high recall indicates that the model successfully identifies the majority of fraudulent providers, missing only a small number of fraud cases.

For non-fraud providers:

- Precision: 99%
- Recall: 91%
- F1 Score: 95%

The model performs exceptionally well in identifying legitimate providers while maintaining strong fraud detection capability.

Overall, Logistic Regression provides a strong baseline model with excellent ROC-AUC performance and very high fraud recall.

## ROC Curve - Logistic Regression

```python
from sklearn.metrics import roc_curve
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_lr)
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=fpr_lr,
        y=tpr_lr,
        mode="lines",
        name=f"Logistic Regression (AUC = {lr_auc:.3f})"
    )
)
fig.add_trace(
    go.Scatter(
        x=[0, 1],
        y=[0, 1],
        mode="lines",
        name="Random Classifier",
        line=dict(dash="dash")
    )
)
fig.update_layout(
    title="ROC Curve - Logistic Regression",
    xaxis_title="False Positive Rate",
    yaxis_title="True Positive Rate",
    width=900,
    height=600
)
fig.show()
```

### Interpretation

The ROC curve demonstrates excellent classification performance.

The Logistic Regression model achieved an ROC-AUC score of approximately 0.97, indicating a very strong ability to distinguish fraudulent providers from non-fraudulent providers.

The curve remains close to the upper-left corner of the graph, significantly outperforming the random classifier baseline.

An ROC-AUC value above 0.90 is generally considered excellent, suggesting that Logistic Regression is already a highly effective fraud detection model for this dataset.

## Random Forest Classifier

Random Forest is an ensemble learning algorithm that combines multiple decision trees to improve predictive performance.

Advantages include:

- Captures non-linear relationships
- Handles complex feature interactions
- Less sensitive to outliers
- Provides feature importance scores
- Often performs exceptionally well on fraud detection problems

```python
rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train, y_train)
```

### Interpretation

The Random Forest model was successfully trained using the balanced training dataset.

Unlike Logistic Regression, Random Forest can capture complex non-linear relationships and feature interactions. This often leads to improved fraud detection performance, particularly when fraudulent behavior patterns are not linearly separable.

## Generate Random Forest Predictions

```python
y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]
print("Random Forest Predictions Generated Successfully")
```

## Evaluate Random Forest Performance

```python
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_precision = precision_score(y_test, y_pred_rf)
rf_recall = recall_score(y_test, y_pred_rf)
rf_f1 = f1_score(y_test, y_pred_rf)
rf_auc = roc_auc_score(y_test, y_prob_rf)
rf_results = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC"],
    "Value": [rf_accuracy, rf_precision, rf_recall, rf_f1, rf_auc]})
rf_results
```

### Interpretation

The Random Forest model achieved strong predictive performance and improved overall classification accuracy compared to Logistic Regression.

Key observations:

- Accuracy increased from 90.57% to 92.33%.
- Precision increased from 49.73% to 56.08%.
- F1 Score improved from 64.34% to 66.67%.

However, the recall score decreased from 91.09% to 82.18%, meaning the model missed more fraudulent providers than Logistic Regression.

This highlights the tradeoff between precision and recall. Random Forest generates fewer false alarms but also fails to identify a greater number of fraud cases.

The choice between models will depend on whether maximizing fraud detection (recall) or minimizing false positives (precision) is the primary business objective.

## Random Forest Confusion Matrix

```python
rf_cm = confusion_matrix(y_test, y_pred_rf)
rf_cm_df = pd.DataFrame(rf_cm,
    index=["Actual Non-Fraud", "Actual Fraud"],
    columns=["Predicted Non-Fraud", "Predicted Fraud"]
)
rf_cm_df
```

### Interpretation

The Random Forest model correctly classified most providers and reduced the number of false positives compared to Logistic Regression.

Results:

- True Negatives (TN): 916
- False Positives (FP): 65
- False Negatives (FN): 18
- True Positives (TP): 83

Compared to Logistic Regression, Random Forest generated fewer false fraud alerts but missed more fraudulent providers.

This explains the higher precision and lower recall observed in the evaluation metrics. The model is more conservative when labeling providers as fraudulent, reducing false alarms while increasing the risk of missing actual fraud cases.

## Visualize Random Forest Confusion Matrix

```python
fig = px.imshow(
    rf_cm,
    text_auto=True,
    color_continuous_scale="Greens",
    labels=dict(
        x="Predicted",
        y="Actual",
        color="Count"
    ),
    x=["Non-Fraud", "Fraud"],
    y=["Non-Fraud", "Fraud"],
    title="Random Forest Confusion Matrix"
)
fig.update_layout( width=700, height=600)
fig.show()
```

### Interpretation

The Random Forest confusion matrix shows strong overall classification performance.

The model correctly identified most non-fraudulent providers while maintaining good fraud detection capability. Compared to Logistic Regression, the Random Forest model produces fewer false fraud alerts but misses a larger number of fraudulent providers.

This behavior reflects the higher precision and lower recall observed in the model evaluation metrics.

### Interpretation

The Random Forest feature importance analysis identifies the variables that contribute most to fraud prediction.

Top fraud indicators include:

2. InpatientClaimCount

4. UniqueInpatientBeneficiaries

The results suggest that unusually high healthcare utilization and reimbursement patterns are among the strongest indicators of potential provider fraud.

## Visualize Feature Importance

```python
top_features = feature_importance.head(10)
fig = px.bar(
    top_features.sort_values(by="Importance"),
    x="Importance",
    y="Feature",
    orientation="h",
    title="Top 10 Random Forest Feature Importances",
    text="Importance")
fig.update_traces(
    texttemplate="%{text:.3f}",
    textposition="outside")
fig.update_layout(width=1000, height=600)
fig.show()
```

### Interpretation

The feature importance visualization highlights the primary drivers of fraud prediction within the Random Forest model.

The most influential variable is MaxLengthOfStay, followed by InpatientClaimCount, TotalInpatientReimbursement, and UniqueInpatientBeneficiaries.

These findings reinforce the exploratory analysis conducted earlier in the project, where fraudulent providers demonstrated substantially higher inpatient activity, longer hospital stays, and larger reimbursement amounts.

The consistency between EDA findings and machine learning feature importance increases confidence in the model's ability to identify meaningful fraud-related patterns.
